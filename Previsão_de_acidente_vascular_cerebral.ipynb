```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV
```

```python
mod = dados[['bmi_label','glucose_label','age_level','work_type','smoking_status']]
```

```python
modififed = pd.get_dummies(mod)
```

```python
modififed.head()
```

```python
modififed.drop(columns=['bmi_label_extreme obese','glucose_label_concerning level',
                        'age_level_super senior citizen','work_type',
                       'smoking_status'],axis=1,inplace=True)
```

```python
dados.drop(columns=['bmi_label','glucose_label','age_level','work_type','smoking_status',
                     'age','avg_glucose_level','bmi'],axis=1,inplace=True)
```

```python
dados.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dados['gender'] = le.fit_transform(dados.gender)
dados['Residence_type'] = le.fit_transform(dados['Residence_type'])
dados['ever_married'] = le.fit_transform(dados['ever_married'])
```

```python
dados = pd.concat([dados,modififed], axis=1)
dados.head()
```

```python
features = dados.drop(columns='stroke',axis=1)
label = dados['stroke']
```

```python
model_params = {
    'random_forest':{
        'models': RandomForestClassifier(),
        'params': {
            'n_estimators': [40,60,100],
            'criterion':['gini','entropy']
        }
    },
    'BernoulliNB':{
        'models': BernoulliNB(),
        'params': {
            'fit_prior': [True,False]
        }
    }
}
```

```python
scores = []

for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['models'], mp['params'], cv=5, return_train_score = False)
    clf.fit(features,label)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_param': clf.best_params_
    })
```

```python
Best= pd.DataFrame(scores)
Best
```

```python
x=dados.drop('stroke',axis=1)
y=dados['stroke']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
cla=RandomForestClassifier()
cla.fit(x_train,y_train)
```

```python
y_pred=cla.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
