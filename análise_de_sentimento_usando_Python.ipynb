# Flipkart analisa an√°lise de sentimento usando Python

O conjunto de dados que estou usando aqui para a an√°lise de sentimentos das avalia√ß√µes do Flipkart √© baixado do Kaggle. Vamos come√ßar esta tarefa importando as bibliotecas Python necess√°rias e o conjunto de dados:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/flipkart_reviews.csv")
print(data.head())
```

Este conjunto de dados cont√©m apenas tr√™s colunas. Vamos dar uma olhada se alguma dessas colunas cont√©m valores ausentes ou n√£o:

```python
print(data.isnull().sum())
```

Portanto, o conjunto de dados n√£o possui nenhum valor nulo. Como esta √© a tarefa da an√°lise de sentimento das avalia√ß√µes do Flipkart, irei limpar e preparar a coluna que cont√©m as avalia√ß√µes antes de ir para a an√°lise de sentimento:

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
```

# An√°lise de sentimento das avalia√ß√µes do Flipkart

A coluna Classifica√ß√£o dos dados cont√©m as classifica√ß√µes dadas por cada revisor. Ent√£o, vamos dar uma olhada em como a maioria das pessoas classifica os produtos que compram na Flipkart:

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
data["Review"] = data["Review"].apply(clean)
```

```python
ratings = data["Rating"].value_counts()
numbers = ratings.index
quantity = ratings.values

import plotly.express as px
figure = px.pie(data, 
             values=quantity, 
             names=numbers,hole = 0.5)
figure.show()
```

Portanto, 60% dos revisores deram 5 de 5 classifica√ß√µes aos produtos que compram da Flipkart. Agora vamos dar uma olhada no tipo de coment√°rios que as pessoas deixam. Para isso, usarei uma nuvem de palavras para visualizar as palavras mais usadas na coluna de resenhas:

```python
text = " ".join(i for i in data.Review)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, 
                      background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Agora analisarei os sentimentos das avalia√ß√µes do Flipkart adicionando tr√™s colunas neste conjunto de dados como Positivo, Negativo e Neutro, calculando as pontua√ß√µes de sentimento das avalia√ß√µes:

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["Review"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["Review"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["Review"]]
data = data[["Review", "Positive", "Negative", "Neutral"]]
print(data.head())
```

Agora vamos ver como a maioria dos revisores pensa sobre os produtos e servi√ßos da Flipkart:

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive üòä ")
    elif (b>a) and (b>c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")
sentiment_score(x, y, z)
```

Portanto, a maioria dos coment√°rios √© neutra. Vamos dar uma olhada no total de pontua√ß√µes de sentimentos positivos, negativos e neutros para encontrar uma conclus√£o sobre as avalia√ß√µes do Flipkart

```python
print("Positive: ", x)
print("Negative: ", y)
print("Neutral: ", z)
```
